```python
import requests
import json
import urllib.request
import time
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
from urllib.request import urlopen
import os
from datetime import date
import re
import json
birth_date = []
died_date = []
birth_place= []
death_place = []
posthumus = []
path="C:/Users/Supra/Documents/Supravat/Documents/ISB/Data Collection"
os.chdir(path)
```

```python
#Do the data cleansing operation on Birth Date and split the Data into Date of Birth and Place of Birth
#fields
```

```python
def process_born_fields(pd_padmadf) :

    for born_field in pd_padmadf['Born'] :
            first_line= False
            second_line= False
            
            try:
                birthdate= born_field.split('\n' ,1)[0]
                if (bool(re.search(r'\d', birthdate))) == True :
               
                    birth_date.append(birthdate)
                    first_line = True
                else :
                  birthdate= born_field.split('\n' ,1)[1].split('\n')[0]
                  if (bool(re.search(r'\d', birthdate))) == True :
                    birth_date.append(birthdate)
                    second_line= True
                  else :
                   birthdate= born_field.split('\n' ,1)[2].split('\n')[0]
                   if (bool(re.search(r'\d', birthdate))) == True :
                    birth_date.append(birthdate) 
            
            except Exception as E :
                 birth_date.append('n/a')
            
            try :
                if first_line == True :
                    birth_place.append(born_field.split('\n' ,1)[1])
                    
                elif second_line == True :
                    birth_place.append(born_field.split('\n' ,1)[2])
                    
                else :
                   
                    birth_place.append(born_field.split('\n' ,1)[3])
                    print(born_field.split('\n' ,1)[3])
            except Exception as E :
                birth_place.append('n/a')
```

```python
#Do the data cleansing operation on Date of Pass Awary and split the Data into Date of Birth
# and Place of Birth fields
```

```python
def process_died_fields(pd_padmadf) :
   

    for died_field in pd_padmadf['Died'] :
            first_line= False
            second_line= False
            
            try:
                deathdate= died_field.split('\n' ,1)[0]
                if (bool(re.search(r'\d', deathdate))) == True :
               
                    died_date.append(deathdate)
                    first_line = True
                else :
                  dieddate= died_field.split('\n' ,1)[1].split('\n')[0]
                  if (bool(re.search(r'\d', dieddate))) == True :
                    died_date.append(deathdate)
                    second_line= True
                  else :
                   deathdate= died_field.split('\n' ,1)[2].split('\n')[0]
                   if (bool(re.search(r'\d', deathdate))) == True :
                    died_date.append(deathdate) 
            
            except Exception as E :
                 died_date.append('n/a')
            
            try :
                if first_line == True :
                    death_place.append(died_field.split('\n' ,1)[1])
                    
                elif second_line == True :
                    death_place.append(died_field.split('\n' ,1)[2])
                    
                else :
                   
                    death_place.append(died_field.split('\n' ,1)[3])
                    print(died_field.split('\n' ,1)[3])
            except Exception as E :
                death_place.append('n/a')
    print(len(died_date))  
    print(len(death_place))
```

```python
# Derive if the award is given posthumusly. We observed that all the recipient who have been given
#awards have a bracket at the end of the name for further reference. We have checked all the data 
# and came into a conclusion that the observation yields correct results.
```

```python
def process_posthumus(pd_padmadf) :
    sub = '['
    for recipient in pd_padmadf['Recipients'] :
         pos = recipient.find(sub)
         if pos > 0 :
            posthumus.append('Yes')
         else :
            posthumus.append('No')
    print(len(posthumus))
    
```

```python
#Read the CSV of the awardee details and insert the fields which have been derived after cleansing records
#Final output has been kept in CSV and json format.
```

```python
def process_csv(filename):
    pd_padmadf = pd.read_csv(filename)
    process_born_fields(pd_padmadf)
    process_died_fields(pd_padmadf)
    csvpos = filename.find('.csv')
    jsonfilename = filename[ :csvpos]+'.json'
    #pd_padmadf['Birth Date'] = birth_date
    process_posthumus(pd_padmadf)
    try :
        pd_padmadf.insert(7, 'Birth Date',  birth_date)
        pd_padmadf.insert(8,'Place of Birth', value = birth_place )
        pd_padmadf.insert(10,'Death Date', value = died_date )
        pd_padmadf.insert(11,'Place of Death' ,value = death_place)
        pd_padmadf['Posthumus'] = posthumus
        pd_padmadf.to_csv(filename, index=False)
        pd_padmadf.to_json(jsonfilename, orient='table')

    except Exception as E:
        print(E)
    
```
